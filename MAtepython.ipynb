```python
import sympy as sp
from sympy.core.symbol import symbols
#from sympy import init_session
x = sp.symbols('x')
y = sp.symbols('y')
#letras
a = sp.symbols('a')
pi = symbols('pi',real = True)
sp.sqrt(a+pi)
```

```python
#Derivada de una función con una variable
f1 = sp.ln(sp.cos(sp.sin(x)))
derivada = sp.diff(f1)
print(f1)
print()
print(derivada)
```

```python
# Derivadas parciales

f2 = x**3*y+y**2
print(f2)
print()
print("La derivada respecto a x\n")
derivadax = sp.diff(f2,x)
print()
print(derivadax)
print()
derivaday = sp.diff(f2,y)
print()
print(derivaday)
```

```python
# Integrales

f3 = 2*x*sp.sqrt(x**3)
sp.integrate(f3,x)
```

```python
# Integrales para más variables
f4 = x*y
sp.integrate(f3,x,y)
```

```python
#Integrales definidas

b = sp.symbols('b')

f5 = sp.cos(x+2*pi)
sp.integrate(f5,(x,a,b))
```

```python
# Integral de Gauss
f6 = sp.exp(-x)*x**(-1/2)
sp.integrate(f6,(x,0,np.infty))
```

```python
# Creación de matrices

A = sp.Matrix([[4,5],[2,-5]])
print(A)
print()
# Determinante de A
print(A.det())
print()
# Matriz escalonada reducida
print(A.rref())
print()
#Eigenvalores y Eigenvectores
print(A.eigenvals())
print()
print(A.eigenvects())
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#Rango y puntos
x = np.linspace(-10,10,100)

#Definir la funcion 

fx= x**2

#imprimir la funcion evaluada en los puntitos
figura,grafica = plt.subplots()
grafica.plot(x,fx)
```

```python
#Grafica de ventas
x = np.arange(0.,2.,.05)
y1 = x

y2 = x**2

plt.plot(x,y1,'.b',label='Lineal',markersize = 2,linewidth=2)
plt.plot(x,y2,'--r',label='Cuadrada',linewidth=2)

# Leyenda a las gráficas
plt.title('Grafica de ventas')
plt.xlabel('Ventas')
plt.ylabel('Ganancias')

plt.legend()#Nos muestra la simbología
plt.grid(True)## Muestra la cuadricula 
plt.show()
```

```python
#Grafica de funciones trigonométricas
#Sin 
x = np.linspace(-np.pi,np.pi,100)

y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x,y1,'-g',label='Coseno',linewidth=1)
plt.plot(x,y2,'*m',label='Seno',linewidth=1)

# Los nombres para x y 'y'
#Gráfica de frecuencias
plt.title('Grafica de frecuencias')
plt.xlabel('Tiempo(s)')
plt.ylabel('Metros(m)')

plt.legend()
plt.grid(True)
plt.show()
```

```python
rango = np.arange(0.,4.0,0.02)
tiempo = np.arange(0.,4.0,0.02)
A = 0.5
lad = 100
T = np.pi/2
phi = 0.5

eo = A*np.cos(np.abs((2*np.pi*rango)/lad+(2*np.pi*tiempo)/T+phi))

figura, grafica = plt.subplots()
grafica.plot(rango,eo)
grafica.set(xlabel='Tiempo(s)',ylabel = 'Distancia(m)',title = 'Ecuacion de onda')
grafica.grid(True)

#Para guardar la gráfica
figura.savefig('Ondas.png')
plt.show()
```
